```python
import numpy as np
import pandas as pd
import seaborn as sns

import gipsy
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.ticker as ticker
```

```python
x = 12
y = 10


pH=7.4
#Insert uniprot Taxonomy ID
taxonomyID=83333
taxonomy_name="Escherichia coli \(strain K12"

csv_folder="Raw"
cut_folder="Cuts"
dbfile="Ecoli_Hres_db.csv"
```

```python
df=pd.read_csv(dbfile)
df_ribosomal=df[df["prot_desc"].str.contains("50S ribosomal protein",case=False)].reset_index()
df_ribosomal_coord=df_ribosomal.loc[:,'A1':]
df_ribosomal_coord.sum()
```

```python
letterindex=gipsy.IndexMaker(x,'l')                                                              
numbercols=np.arange(1,y+1)                                                                
sns.set(style="white")                                                                     


fig,ax = plt.subplots()
rsdata=df_ribosomal_coord.sum().values.reshape((x,y)).T
df_rs=pd.DataFrame(rsdata,columns=letterindex,index=numbercols)
    
sns_plot=sns.heatmap(df_rs,cmap="Blues",annot=True,cbar=True)
sns_plot.xaxis.tick_top()

plt.show()
plt.close()
```
